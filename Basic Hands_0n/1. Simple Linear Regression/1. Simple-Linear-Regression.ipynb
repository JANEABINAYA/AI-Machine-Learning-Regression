```python
#pandas is a library and it contains the functions related to the database
import pandas as pd
```

```python
#dataset is called by read_csv function under pandas and it is assigned to the dataset vareable
dataset = pd.read_csv('6. Salary_Data.csv')
```

```python
#calling out the dataset
dataset 
```

```python
#calling only the input in the dataset by a vareable
independent = dataset[["YearsExperience"]]
independent
```

```python
#calling only the output in the dataset by a vareable 
dependent = dataset[["Salary"]]
dependent
```

```python
#importing the train_test_split from sklearn.model_selection
from sklearn.model_selection import train_test_split
#calling the imported file
X_train, X_test, Y_train, Y_test=train_test_split(independent, dependent,test_size=0.30,random_state=0)
```

```python
#importing the LinearRegression from sklearn.linear_model
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, Y_train)
```

```python
#calculating the weight by coef_
weight=regressor.coef_
weight
```

```python
#calculating the bias by intercept_
bias=regressor.intercept_
bias
```

```python
#assigning a value to a vareable
y_pred=regressor.predict(X_test)
```

```python
#importing r2_score from sklearn.metrics
from sklearn.metrics import r2_score
r_score=r2_score(Y_test, y_pred)
```

```python
#calling for result
r_score
```

```python
#pickle is used To save trained machine learning models
import pickle
filename = 'finalized_model.sav'
```

```python
#assigning the model to the filename
pickle.dump(regressor,open(filename,'wb'))
```

```python
loaded_model = pickle.load(open('finalized_model.sav', 'rb'))
result = loaded_model.predict([[15]])
```
